# IMDB Capstone Project Phase 1

```python
# Importing the required libraries
from selenium import webdriver
import pandas as pd
import numpy as np
from time import sleep
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
from bs4 import BeautifulSoup
import requests
import re
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotInteractableException

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Using selenium I am scraping Sno, URL, Votes, Duration
# opening the Chrome window
driver=webdriver.Chrome('D:\chromedriver.exe')

#maximize the browser
driver.maximize_window()

#Opening the website
driver.get("https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1")


# To scrap URL of the data
Sno=[]
URL=[]
Votes=[]
Duration=[]

for i in range(0,35):
    links=driver.find_elements(By.XPATH, '//h3[@class="lister-item-header"]/a')  
    for i in links:
        URL.append(i.get_attribute('href'))
    
    #Scrapping the Sr. No.
    no=driver.find_elements(By.XPATH, '//span[@class="lister-item-index unbold text-primary"]')
    for j in no:
        Sno.append(j.text.strip())
    
        
    #Scrapping the Votes
    vote=driver.find_elements(By.XPATH, '//p[@class="sort-num_votes-visible"]/span[2]')
    for k in vote:
        Votes.append(k.text.strip())
    
    # Scrapping Duration
    dur=driver.find_elements(By.XPATH,'//span[@class="runtime"]')
    for l in dur:
        Duration.append(l.text.replace("    min",""))
      
          
    try:
        next_but=driver.find_element(By.XPATH,'(//a[@class="lister-page-next next-page"])[2]')
        next_but.click()
        sleep(4)
    except NoSuchElementException:
        pass

```

```python
len(URL)
```

```python
# Creating Empty lists
Movie_Name=[]
Director_Name=[]
Year=[]
Ratings=[]
Metascore=[]
Genre=[]
Popularity=[]
Gross=[]
Certification=[]
Director1=[]
Director2=[]
Director3=[]
Star1=[]
Star2=[]
Star3=[]
Star4=[]
```

```python
# for loop for the information to be scrapped
headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36"}

for i in URL:
    response=requests.get(i,headers=headers)
    soup=BeautifulSoup(response.content,'html.parser')
    
    #Scraping Movie_Name
    MN=soup.find('span',class_='sc-afe43def-1 fDTGTb') if soup.find('span',class_='sc-afe43def-1 fDTGTb') else ""
    Movie_Name.append(MN.get_text().strip())
    
    #Scraping Year
    year=soup.find('div',class_='sc-52d569c6-0 kNzJA-D').ul.li.a if soup.find('div',class_='sc-52d569c6-0 kNzJA-D').ul.li.a else ""
    Year.append(year.get_text().strip())
    
    #Scraping Certification
    try:
        certi=soup.find('div',class_='sc-52d569c6-0 kNzJA-D').ul.find_all('li',class_="ipc-inline-list__item")[1].a if soup.find('div',class_='sc-52d569c6-0 kNzJA-D').ul.find_all('li',class_="ipc-inline-list__item")[1].a else ""
        Certification.append(certi.get_text())
    except:
        Certification.append('NaN')
    
    #Scraping Genre
    genre=soup.find('div',class_='ipc-chip-list__scroller') if soup.find('div',class_='ipc-chip-list__scroller') else ""
    Genre.append(genre.get_text())
    
    #Scraping Ratings
    try:
        rate=soup.find('span',class_='sc-bde20123-1 iZlgcd') if soup.find('span',class_='sc-bde20123-1 iZlgcd') else ""
        Ratings.append(rate.get_text())
    except:
        Ratings.append("NaN")
    
    #Scraping Metascore
    try:
        meta=soup.find('span',class_='score-meta') if soup.find('span',class_='score-meta') else ""
        Metascore.append(meta.get_text())
    except:
        Metascore.append("NaN")
    
    #Scraping Popularity
    try:
        popu=soup.find('div',class_='sc-5f7fb5b4-1 bhuIgW') if soup.find('div',class_='sc-5f7fb5b4-1 bhuIgW') else ""
        Popularity.append(popu.get_text())
    except:
        Popularity.append("NaN")
    
    #Scraping Gross
    try:
        gross=soup.find('ul',class_='ipc-metadata-list ipc-metadata-list--dividers-none ipc-metadata-list--compact sc-6d4f3f8c-0 VdkJY ipc-metadata-list--base')
        p=gross.get_text().replace(",","")
        match=re.findall(r'\d+', p)
        Gross.append(match[-1])
    except:
        Gross.append('NaN')
    
    #Scraping Director 1
    Dire1=soup.find('ul',class_='ipc-metadata-list ipc-metadata-list--dividers-all sc-bfec09a1-8 iiDmgX ipc-metadata-list--base').div.ul.find_all('li',class_='ipc-inline-list__item')[0] if soup.find('ul',class_='ipc-metadata-list ipc-metadata-list--dividers-all sc-bfec09a1-8 iiDmgX ipc-metadata-list--base').div.ul.find_all('li',class_='ipc-inline-list__item')[0] else ""
    Director1.append(Dire1.get_text())
    
    
    #Scraping Director 2
    try:
        Dire2=soup.find('ul',class_='ipc-metadata-list ipc-metadata-list--dividers-all sc-bfec09a1-8 iiDmgX ipc-metadata-list--base').div.ul.find_all('li',class_='ipc-inline-list__item')[1] if soup.find('ul',class_='ipc-metadata-list ipc-metadata-list--dividers-all sc-bfec09a1-8 iiDmgX ipc-metadata-list--base').div.ul.find_all('li',class_='ipc-inline-list__item')[1] else ""
        Director2.append(Dire2.get_text())
    except:
        Director2.append('NaN')
    
    #Scraping Director 3
    try:
        Dire3=soup.find('ul',class_='ipc-metadata-list ipc-metadata-list--dividers-all sc-bfec09a1-8 iiDmgX ipc-metadata-list--base').div.ul.find_all('li',class_='ipc-inline-list__item')[2] if soup.find('ul',class_='ipc-metadata-list ipc-metadata-list--dividers-all sc-bfec09a1-8 iiDmgX ipc-metadata-list--base').div.ul.find_all('li',class_='ipc-inline-list__item')[2] else ""
        Director3.append(Dire3.get_text())
    except:
        Director3.append('NaN')
    
    # Scrapping star1
    star=soup.find_all('div',class_='sc-bfec09a1-5 kUzsHJ')[0].find('a',class_='sc-bfec09a1-1 fUguci')
    Star1.append(star.get_text())
    
    # Scrapping star2
    try:
        star=soup.find_all('div',class_='sc-bfec09a1-5 kUzsHJ')[1].find('a',class_='sc-bfec09a1-1 fUguci')
        Star2.append(star.get_text())
    except:
        Star2.append("NaN")
    
    # Scrapping star3
    try:
        star=soup.find_all('div',class_='sc-bfec09a1-5 kUzsHJ')[2].find('a',class_='sc-bfec09a1-1 fUguci')
        Star3.append(star.get_text())
    except:
        Star3.append('NaN')
    
    # Scrapping star4
    try:
        star=soup.find_all('div',class_='sc-bfec09a1-5 kUzsHJ')[3].find('a',class_='sc-bfec09a1-1 fUguci')
        Star4.append(star.get_text())
    except:
        Star4.append('NaN')
    
    
    
    
```

```python
# making a dataframe 1 and saving to csv
dataset1=pd.DataFrame({'Sno':Sno,'Movie_Name':Movie_Name,'Director1':Director1,'Director2':Director2,'Director3':Director3,'Duration':Duration,'Year':Year,'Ratings':Ratings,'Metascore':Metascore})

print(dataset1)

# To CSV
dataset1.to_csv('table1.csv')
```

```python
# making a dataframe 2 and saving to csv
dataset2=pd.DataFrame({'Movie_Name':Movie_Name,'Star1':Star1,'Star2':Star2,'Star3':Star3,'Star4':Star4,'Votes':Votes,'Genre':Genre,'Gross_Collection':Gross,'Popularity':Popularity,'Certification':Certification})

print(dataset2)

# To CSV
dataset2.to_csv('table2.csv')
```

```python
#Importing libraries Required
import pandas as pd
import numpy as np
import re
import sqlite3
```

```python
# Making Connection in the database
db=sqlite3.connect("IMDB2.db")

# Creating the Cursor object to execute the query
cursor = db.cursor()
```

```python
# load data from table 1 to df1
df1 = pd.read_csv('table1.csv',index_col=['Sno'])
df1.drop('Unnamed: 0',axis=1)
```

```python
# load data from table 2 to df2
df2 = pd.read_csv('table2.csv',index_col=[0])
df2
```

```python
# Checking the data types of the table1
df1.info()
```

Table 1 data types are correct ones

```python
# Checking data types of the table 2
df2.info()
```

It can be seen that Popularity and Votes are Numeric columns being treated as object has to be treated

```python
# Converting the data type of datatype of columns Votes and Popularity
df2['Votes']=df2['Votes'].str.replace(",","")
df2['Votes']=pd.to_numeric(df2['Votes'])
df2['Popularity']=df2['Popularity'].str.replace(",","")
df2['Popularity']=pd.to_numeric(df2['Popularity'])

df2.dtypes
```

```python
# Splitting the Genre Column in genre1, genre2, genre3
list1=[]
for i in df2['Genre']:
    z=re.findall('[A-Z][^A-Z]*', i)
    list1.append(z)
    
df2['genre']=list1

df2['genre1'] = df2['genre'].str.get(0)
df2['genre2'] = df2['genre'].str.get(1)
df2['genre4'] = df2['genre'].str[2]
df2['genre5'] = df2['genre'].str[3]

df2['genre3'] = df2[['genre4', 'genre5']].stack().groupby(level=0).agg(''.join)

# Droping the unneccesary columns
df2=df2.drop(['genre','Genre','genre4','genre5'],axis=1)
df2.head(5)
```

```python
# drop data into database
df1.to_sql("Table1", db)

df2.to_sql("Table2", db)
```
